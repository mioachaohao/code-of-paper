```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path ="Descriptive_feature.csv"
data = pd.read_csv(file_path)

# Set the figure size and style
plt.figure(figsize=(10, 5))
plt.style.use('seaborn-darkgrid')

# Set the color palette
colors = sns.color_palette('Set3', 3)

# Create a twin Axes sharing the x-axis
ax1 = plt.gca()
ax2 = ax1.twinx()

# Plotting the histograms for upstream_nodes and upstream_edges
ax1.bar(data['years'] - 0.2, data['upstream_nodes'], width=0.4, label='Upstream Nodes', alpha=0.6, color=colors[0])
ax1.bar(data['years'] + 0.2, data['upstream_edges'], width=0.4, label='Upstream Edges', alpha=0.6, color=colors[1])

# Plotting the line plot for upstream_density
ax2.plot(data['years'], data['upstream_density'], label='Upstream Density', marker='o', color=colors[2], linestyle='-', linewidth=2)

# Setting the limit for the right y-axis
ax2.set_ylim(0, 0.04)

# Adding labels
ax1.set_xlabel('Year')
ax1.set_ylabel('Count')
ax2.set_ylabel('Density')



# Merge legends
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines = lines1 + lines2
labels = labels1 + labels2

# Create a merged legend
ax1.legend(lines, labels, loc='upper left')

# Remove title
plt.title('')

# Adding grid
ax1.grid(True)
ax2.grid(False)

# Set resolution to 300 DPI and save the image
plt.savefig("results/figure 4.png", dpi=300)

# Show the plot
plt.show()
```

```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

# File path
nodelist_file_path = "wb_nodelist.xlsx"

# Load node list data for the years 1995, 2010, 2019, and 2007 from different worksheets in the Excel file
nodelist_1995 = pd.read_excel(nodelist_file_path, sheet_name='nodelist_1995')
nodelist_2010 = pd.read_excel(nodelist_file_path, sheet_name='nodelist_2010')
nodelist_2019 = pd.read_excel(nodelist_file_path, sheet_name='nodelist_2019')
nodelist_2007 = pd.read_excel(nodelist_file_path, sheet_name='nodelist_2007')
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import geopandas as gpd
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

def plot_world_map_viridis(world, nodelist, ax, title):
    # Merge node data into the world map dataframe
    world = world.merge(nodelist, how="left", left_on="iso_a3", right_on="Id")
    
    # Get unique values of communities and color mapping
    unique_communities = world['community'].dropna().unique()
    unique_communities.sort()
    
    colors = sns.color_palette("Set3", len(unique_communities))
    
    # Plot the map
    world.boundary.plot(ax=ax, linewidth=1)
    
    for i, community in enumerate(unique_communities):
        world[world['community'] == community].plot(color=colors[i], linewidth=0.8, ax=ax, edgecolor='0.8')
    
    # Create a discrete legend
    legend_labels = [mpatches.Patch(color=colors[i], label=f'Community {int(community)}') for i, community in enumerate(unique_communities)]
    ax.legend(handles=legend_labels, title="Communities", bbox_to_anchor=(1, 1), loc='upper left')
    
    # Set the title
    ax.set_title(title)

    # Remove axes
    ax.axis('off')

# Create a 2x2 figure
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Adjust horizontal and vertical spacing between subplots
plt.subplots_adjust(wspace=0.4, hspace=-0.3)

# Plot the world map for 1995 using the viridis color scheme and discrete legend
plot_world_map_viridis(world, nodelist_1995, axes[0, 0], "1995")

# Plot the world map for 2007 using the viridis color scheme and discrete legend
plot_world_map_viridis(world, nodelist_2007, axes[0, 1], "2007")

# Plot the world map for 2010 using the viridis color scheme and discrete legend
plot_world_map_viridis(world, nodelist_2010, axes[1, 0], "2010")

# Plot the world map for 2019 using the viridis color scheme and discrete legend
plot_world_map_viridis(world, nodelist_2019, axes[1, 1], "2019")

# Save the composite image
plt.savefig("results/figure 6.png", dpi=300)

# Display the image
plt.show()
```

```python
import pandas as pd
import networkx as nx
from collections import defaultdict
from networkx.algorithms import community as nx_comm

import os

# Get the current working directory
current_directory = os.getcwd()

# Construct the absolute file path
file_path = os.path.join(current_directory, "wb_edgelist.xlsx")

# Load the file using the absolute path
edges_data = {sheet_name: pd.read_excel(file_path, sheet_name=sheet_name) for sheet_name in pd.ExcelFile(file_path).sheet_names}

# Build networks for each year
graphs = {}
for year, edges_df in edges_data.items():
    G = nx.from_pandas_edgelist(edges_df, 'Source', 'Target', ['weight'], create_using=nx.DiGraph)
    graphs[year] = G

# Load node data (all tables)
nodes_dfs = pd.read_excel("wb_nodelist.xlsx", sheet_name=None)

# Initialize a dictionary to store modularity results for each year
modularity_results_all_years = {}

# Calculate modularity for each year using corresponding community information
for year, G in graphs.items():
    # Load the node data corresponding to that year
    nodes_df_year = nodes_dfs.get('nodelist_' + year.split('_')[-1], None)
    
    # If node data doesn't exist, skip that year
    if nodes_df_year is None:
        print(f"No node list data available for {year}, skipping...")
        continue
    
    # Extract community information from node data
    community_mapping = nodes_df_year.set_index('Id')['community'].to_dict()
    communities_dict = defaultdict(list)
    for node, community in community_mapping.items():
        if node in G.nodes:
            communities_dict[community].append(node)
    communities_list = list(communities_dict.values())
    
    # Calculate modularity
    modularity = nx_comm.modularity(G, communities_list)
    modularity_results_all_years[year] = modularity

# Print modularity results
print(modularity_results_all_years)
```
